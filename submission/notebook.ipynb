## Exploratory Data Analysis (EDA)

### Explore Data orders_dataset.csv dan products_dataset.csv

```python
# Analisis satu
# Mengubah kolom order_purchase_timestamp menjadi format datetime
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'])

# Menghitung jumlah pembelian harian
daily_purchase_counts = orders_df['order_purchase_timestamp'].dt.date.value_counts().sort_index().reset_index()
daily_purchase_counts.columns = ['purchase_date', 'purchase_count']

# Menemukan tanggal dengan pembeli terbanyak
max_purchase_date = daily_purchase_counts.loc[daily_purchase_counts['purchase_count'].idxmax()]['purchase_date']
max_purchase_count = daily_purchase_counts['purchase_count'].max()


print(f'Tanggal dengan Pembelian Terbanyak: {max_purchase_date}')
print(f'Jumlah Pembelian: {max_purchase_count}')
print("\n")

# Anlisis kedua
# Hitung jumlah produk yang ada disetiap kategori
category_counts = products_df['product_category_name'].value_counts()

# Menemukan kategori terbanyak
max_category = category_counts.idxmax()
max_category_count = category_counts.max()

print(f'Kategori dengan Produk Terbanyak: {max_category}')
print(f'Jumlah Produk: {max_category_count}')

# Bikin dataframe baru untuk menyimpan data
main_df = pd.DataFrame({
    'order_purchase_timestamp': orders_df['order_purchase_timestamp'],
    'product_category_name': products_df['product_category_name'],

})

# Convert dataframe ke csv untuk streamlit di visual studio :D
main_df.to_csv("main_data.csv", index=False)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.figure(figsize=(12, 6))
plt.bar(daily_purchase_counts['purchase_date'], daily_purchase_counts['purchase_count'], color='blue')
plt.title('Jumlah Pembelian per Tanggal')
plt.xlabel('Tanggal Pembelian')
plt.ylabel('Jumlah Pembelian')
plt.xticks(rotation=45)
plt.show()
```

### Pertanyaan 2:

```python
category_counts = products_df['product_category_name'].value_counts()

max_category = category_counts.idxmax()
max_category_count = category_counts.max()

plt.figure(figsize=(12, 6))
bars = plt.bar(category_counts.index, category_counts, color='green')
plt.title('Jumlah Produk per Kategori')
plt.xlabel('Kategori Produk')
plt.ylabel('Jumlah Produk')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
